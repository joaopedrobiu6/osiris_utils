```python
import sys
from pathlib import Path

import matplotlib.pyplot as plt

import osiris_utils as ou
```

```python
sim_path = "example_data/thermal.1d"
sim = ou.Simulation(sim_path)
```

```python
# grab Ez diagnostic
ez = sim["e3"]
```

```python
# plot Ez field at iteration 10
plt.plot(ez.x, ez[10], label=f"${ez.label}$")
plt.title(rf"${ez.label}$ at t = {ez.time(10)[0]} $[{ez.time(10)[1]}]$")
plt.xlabel(ez.axis[0]["plot_label"])
plt.ylabel(f"${ez.units}$")
plt.legend()
plt.tight_layout()
plt.show()
```
